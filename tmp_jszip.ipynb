# Download JSZip v3.10.1 Minified File
Use Python to fetch the minified build from the CDN.

```python
import urllib.request

JSZIP_URL = "https://cdn.jsdelivr.net/npm/jszip@3.10.1/dist/jszip.min.js"

with urllib.request.urlopen(JSZIP_URL) as response:
    jszip_bytes = response.read()
```

# Verify Download Size and Hash
Compute size and SHA-256 to confirm integrity.

```python
import hashlib

length = len(jszip_bytes)
sha256 = hashlib.sha256(jszip_bytes).hexdigest()
print(f"bytes: {length}")
print(f"sha256: {sha256}")
```

<VSCode.Cell language="markdown">
# Save Local Vendor Copy
Write the downloaded bytes to a local jszip.min.js file.
</VSCode.Cell>

```python
<VSCode.Cell language="python">
output_path = "jszip.min.js"
with open(output_path, "wb") as fh:
    fh.write(jszip_bytes)
print(f"saved to {output_path}")
</VSCode.Cell>
```

<VSCode.Cell language="markdown">
# Output Minified Contents
Print the full JSZip minified source for direct vendoring.
</VSCode.Cell>

```python
<VSCode.Cell language="python">
import sys

jszip_text = jszip_bytes.decode("utf-8")
sys.stdout.write(jszip_text)
</VSCode.Cell>
```
